# KDE plot

```python
import seaborn as sns
```

```python
data = np.random.normal(size=3000)
sns.kdeplot(data);
```

```python
data = np.random.multivariate_normal([2,2],[[1,0.5],[0.5,1]],5000)
sns.kdeplot(*data.T, color="k");
```

```python
sns.kdeplot(*data.T, shade=True, shade_lowest=False);
```

```python
sns.kdeplot(*data.T, 
            cmap=sns.cubehelix_palette(light=1, as_cmap=True),
            shade=True,
            n_levels=10);
```
