# Plot

Plot a list of $y$-s

```python
data_list = np.arange(5,10,0.1)
plt.plot(data_list,'.');
```

Plot multiple datasets stored in an array. If array is oriented in the way it would make sense, then it has to be transposed.

```python
data = np.array([i*np.arange(5,10,0.1) for i in range(3)])
plt.plot(data.T,'.');
```

```python
data = [i*np.arange(5,10,0.1) for i in range(3)]
[plt.plot(line,'.') for line in data];
```

Plot a vectorised function

```python
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y);
```

```python
x = np.linspace(0,5,100)
data = [np.sin(i * x) for i in range(3)]
[plt.plot(x, line) for line in data];
```

Plot a function

```python
x = np.linspace(0,10,50)
y = list(map(np.sin,x))
plt.plot(x,y);
```

Single line

```python
plot = lambda y,xparams : plt.plot(np.linspace(*xparams),list(map(y,np.linspace(*xparams))))
plot(np.sin,(0,4,100));
```
