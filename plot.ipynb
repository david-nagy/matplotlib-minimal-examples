# Plot

Plot a list of $y$-s

```python
data_list = np.arange(5,10,0.1)
plt.plot(data_list,'.');
```

Plot a function

```python
x = np.linspace(0,10,50)
y = list(map(np.sin,x))
plt.plot(x,y);
```

Plot a vectorised function

```python
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y);
```

Single line

```python
plot = lambda y,xparams : plt.plot(np.linspace(*xparams),list(map(y,np.linspace(*xparams))))
plot(np.sin,(0,4,100));
```
