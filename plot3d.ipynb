```python
t = np.linspace(10,100,100)
t_total = np.linspace(10,100,100)
X, Y = np.meshgrid(t_total,t)
Z = tfp.distributions.Uniform(0,X).prob(Y)
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis');
```

Rotate viewpoint and add axis labels

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis');

ax.view_init(30, 250)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

Wireframe plot

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='viridis')
ax.view_init(90, 270)
ax.set_zticks([]);
```
