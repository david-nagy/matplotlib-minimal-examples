# Subplot

```python
import matplotlib.gridspec as gridspec
subplots = [plt.subplot(grid_loc) for grid_loc in gridspec.GridSpec(5, 10)]
[(sp.plot(np.random.rand(10)), sp.axis('off')) for sp in subplots]
plt.show();
```

```python
n,m = 2,2
plt.subplot(n,m,1)
plt.plot(range(12))
plt.axis('off')
plt.subplot(n,m,2)
plt.plot(np.sin(np.linspace(0,5)))
plt.axis('off')
plt.show()
```
